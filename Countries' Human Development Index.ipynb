```python
# ===  Step 2: Bottom 10 HDI for African countries ===
bottom10 = df.nsmallest(10, "HDI")
print("=== Bottom 10 African Countries by HDI ===")
print(bottom10[["Country", "HDI", "Life Expectancy", "Expected Schooling", "Mean Schooling", "GNI per Capita"]])
```

```python
# === Step 3: Bar chart of bottom 10 HDIs for African countries ===
plt.figure(figsize=(10,6))
plt.barh(bottom10["Country"][::-1], bottom10["HDI"][::-1], color="tomato")
plt.xlabel("HDI (2025)")
plt.title("Bottom 10 African Countries by HDI")
plt.tight_layout()
plt.show()
```

```python
# === Step 4: Scatter plots of HDI vs other indicators ===
indicators = ["Life Expectancy", "Expected Schooling", "Mean Schooling", "GNI per Capita"]

plt.figure(figsize=(14,10))
for i, col in enumerate(indicators, 1):
    plt.subplot(2, 2, i)
    plt.scatter(df[col], df["HDI"], color="skyblue", edgecolor="black")
    plt.xlabel(col)
    plt.ylabel("HDI")
    plt.title(f"HDI vs {col}")
plt.tight_layout()
plt.show()
```

```python
# === Step 5: Correlations ===
print("\n=== Correlations between HDI and other indicators ===")
print(df[["HDI"] + indicators].corr()["HDI"].drop("HDI"))
```
