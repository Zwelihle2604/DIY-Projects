Introduction

This report analyzes load-shedding hours for 2023 and 2024 to assess changes in the frequency and duration of power outages. Load shedding continues to affect productivity, service delivery, and overall energy reliability, making it important to track performance over time.

The analysis compares total and monthly load-shedding hours across both years to identify key trends, peak periods, and improvements or declines in supply stability. The findings aim to support better planning and decision-making within the energy sector.

```python
import pandas as pd
import matplotlib  as plt
import numpy as np
```

```python
#Load the file that contains the data
data = pd.read_csv("Loadshedding hours.csv")
```

```python
#Display the content of the file
data
```

```python
# Display data for 2023 and 2024
df = data[['Month', '2023_hours', '2024_hours']]
df
```

```python
# Create a plot for 2023
plt.figure(figsize=(8, 5))
plt.plot(df["Month"], df["2023_hours"], marker='o', label="2023 Hours", color= 'orange')
plt.title("Monthly Hours: 2023")
plt.xlabel("Month")
plt.ylabel("Hours")
plt.legend()
plt.grid(False)
```

```python
#Create a plot for 2024
plt.figure(figsize=(8, 5))
plt.plot(df["Month"], df["2024_hours"], marker='o', label="2024 Hours")
plt.title("Monthly Hours: 2024")
plt.xlabel("Month")
plt.ylabel("Hours")
plt.legend()
plt.grid(False)
```

Conclusion

The data shows that load shedding in South Africa eased substantially in 2024 compared to 2023, suggesting improved power generation capacity, better management, or reduced demand pressure on the grid.
