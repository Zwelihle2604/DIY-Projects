Introduction

This analysis explores patterns and trends within a movie ratings dataset to uncover insights into audience preferences and film performance. By focusing solely on the ratings data, the objective is to identify which movies achieved the highest viewer satisfaction and which received the lowest. Through sorting, comparison, and visualization of ratings, this project aims to highlight how audiences perceive different films and what separates top-rated titles from those that underperform. The findings provide a foundation for understanding viewer sentiment and can guide future decisions in content production, marketing, and audience targeting.

```python
#Import relevant functional libraries:
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load the csv file to be used:
ratings = pd.read_csv("Movie Ratings.csv", on_bad_lines='skip', engine='python')
```

```python
#Display the dataframe and the content of the file:
ratings
```

```python
#Find the movie with the highest rating
highest_rated_movie = ratings.loc[ratings['rating_val'].idxmax()]

print("Movie with the Highest Rating:")
print(highest_rated_movie['movie_id'])
```

```python
#Find the movie with the lowest rating
lowest_rated_movie = ratings.loc[ratings['rating_val'].idxmin()]
print("Movie with the Lowest Rating:")
print(lowest_rated_movie['movie_id'])
```

```python
#Display the top 10 movies highest rated movies:
top10_movies = ratings.sort_values(by='rating_val', ascending=False).head(10)
print("Top 10 Highest-Rated Movies:")
top10_movies
```

```python
#Display the bottom 10 lowest rated movies:
bottom10_movies = ratings.sort_values(by='rating_val', ascending=False).tail(10)
print("Bottom 10 Lowest-Rated Movies:")
bottom10_movies
```

```python
#Visualise the Top 10 Highest Rated Movies:

#Sort movies by rating in descending order and get top 10
top10_movies = ratings.sort_values(by='rating_val', ascending=False).head(10)

# Plot a bar chart
plt.figure(figsize=(10,6))
plt.barh(top10_movies['movie_id'], top10_movies['rating_val'], color='skyblue')
plt.xlabel('Movie')
plt.ylabel('Rating')
plt.title('Top 10 Highest-Rated Movies')
plt.gca().invert_yaxis()  # Highest rating on top
plt.show()
```

```python
#Visualise the bottom 10 Lowest Rated Movies:
#Sort movies by rating in ascending order and get bottom 10
bottom10_movies = ratings.sort_values(by='rating_val', ascending=True).head(10)


# Plot a bar chart
plt.figure(figsize=(10,6))
plt.barh(bottom10_movies['movie_id'], bottom10_movies['rating_val'], color='salmon')
plt.xlabel('Rating')
plt.ylabel('Movie')
plt.title('Bottom 10 Lowest-Rated Movies')
plt.show()
```

Summary of Movie Ratings Analysis:

The analysis of the dataset reveals clear contrasts in audience reception across movies.

The top 10 highest-rated movies demonstrate strong viewer approval, suggesting high-quality production, engaging storytelling, or strong audience appeal.

In contrast, the bottom 10 lowest-rated movies highlight titles that may have underperformed due to weak scripts, poor direction, or low audience engagement.

Overall, the distribution of ratings provides valuable insight into audience preferences and can help identify patterns that influence a movieâ€™s success.
